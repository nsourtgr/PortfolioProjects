<b>Visualizing Airbnb data of Athens, Greece (tableau)<b>

The dataset used for this project had been downloaded from insideairbnb. Includes listing activities and metrics for Airbnb in Athens, Greece (last data scraped 28 Dec 2022)
Data source http://insideairbnb.com/

Data csv file imported in Excel and as I observed, I had information on the hosts, geographical information and metrics such as price, number of reviews and other related fields.

Next stage was data cleaning. Data had to be cleaned and fix in certain cases cause of extreme price difference (outliers) and null values.

I converted Latitude and Longitude to Address and the info imported in Excel (Applying VBA Code or Using Online Geocoding Tool).
https://www.exceldemy.com/convert-latitude-and-longitude-to-address-in-excel/

Post Code extracted so data to be used for mapping visualization.

Interactive Dashboard
I believe I have made all the charts I need to create an interactive, informative dashboard from the maps and charts I have created about the Airbnb data of Athens.

For this, I make a straight forward dashboard without dazzling visual elements. Just a simple environment to communicate my findings. Here is a screenshot of my resulting dashboard and the link to the dashboard in my Tableau Public account.
